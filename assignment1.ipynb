# Pseudo-Inverse
## c)  Determine P for the data in b) using the pseudo inverse method. 

```python
A_tild = A[:, :-1]
A_tild.shape
```

```python
b = np.array(u_v)
b = b.reshape(-1)
b
```

```python
A_dagger = np.linalg.pinv(A_tild)
p_tild = A_dagger @ b
p_tild
```

```python
p = np.append(p_tild, 1)
P = p.reshape(3, 4)
P
```
